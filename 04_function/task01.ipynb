```python
import re

def validation(raw_username):
    try:
        if not isinstance(raw_username, str):
            print("エラー: ユーザ名は <str> 型で渡してください")
            return False

        # 1. 文字数チェック (5文字以上15文字以内)
        if not (5 <= len(raw_username) <= 15):
            print("エラー: ユーザー名は5文字以上15文字以内で入力してください")
            return False    

        # 2. 空白チェック
        if " " in raw_username:
            print("エラー: ユーザー名に空白を含めることはできません")
            return False
    
        # 3. ユーザ名のチェック
        if re.match(r"^[0-9a-zA-Z]+$", raw_username) is None:
            print("エラー: ユーザー名は英数字のみ使用できます")
            return False
    
    except Exception as e:
        print(f"エラー: 意図しない例外が発生しました", e)
        return False

    else:
        # 上記のチェックや意図しない例外が発生しない場合は True を返す
        return True
    
if validation(input("ユーザ名を入力してください:")):
    print("ユーザー名は有効です")
else:
    print("不正なユーザー名です")
```
