```python
def is_palindrome(s):
    if s == s[::-1]:
        print(f"「{s}」は回文です！")
    else:
        print(f"「{s}」は回文ではないです。")


def greet(name):
    print(f"こんにちは、{name}さん！")


def introduce(name, age):
    print(f"こんにちは、{name}さん！ 年齢は{age}歳ですね。")


def introduce(**kwargs):
    for key, value in kwargs.items():
        print(f"{key} : {value}")


def add(a, b):
    result = a + b
    return result


def calculate(a, b):
    sum_result = a + b
    product_result = a * b
    return sum_result, product_result
```
