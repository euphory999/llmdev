```python
i = 0

while True: # <----------------+
    i += 1                   # | continue ... ループ処理の先端に戻る
                             # |
    if i > 10:               # |
        break # ---------------*------+
                             # |      | break ... 1つループ処理を抜ける
    if i % 2 == 1:           # |      |
        continue # ------------+      |
                                    # |
    print(i)                        # |
# <-----------------------------------+

print("処理を終了します")
```
