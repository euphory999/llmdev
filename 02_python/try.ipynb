```python
try:
    num = int(input("整数を入力してください: "))
    result = 100 / num
    print(f"100 を {num} で割った結果は {result} です")
except ZeroDivisionError:
    print("エラー: 0 で割ることはできません")
except ValueError:
    print("エラー: 整数を入力してください")
except Exception as e:
    # 予期しない例外をキャッチ
    print(f"予期しないエラーが発生しました: {e}")
```
