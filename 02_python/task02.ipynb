```python
import re

## ====================
## 定数
## ====================
# 消費税率(計算の精度については今回は割愛させてください)
CONSUMPTION_TAX_RATE = 0.100

# 送料無料となる基準額
FREE_POSTAGE_PRICE = 2000

# 送料
POSTAGE_FEE = 350

## ====================
## プログラム部
## ====================

# 1. 税抜き価格を教えてもらう (バリデーションやってみよう)
is_validation_ng = True
while is_validation_ng:
    raw_price = input("税抜き価格を教えて下さい")
    is_validation_ng = re.match(r"^[0-9]+$", raw_price) is None
    if is_validation_ng:
        print("税抜き価格は全て半角数字で入力してください!")
price = int(raw_price)

# 2. 税込み価格を計算する (端数切り捨て扱い)
tax_incl_price = int(price * (1.000 + CONSUMPTION_TAX_RATE))

# 3. 送料が必要かどうかを判定して必要ならば加算する
if tax_incl_price >= FREE_POSTAGE_PRICE:
    # 3-a. FREE_POSTAGE_PRICE 円以上ならば送料無料
    print(f"""\
送料は無料です
送料込税込価格は{tax_incl_price}円です""")

else:
    # 3-b. FREE_POSTAGE_PRICE 円未満ならば送料追加
    total_price = int(tax_incl_price + POSTAGE_FEE)
    print(f"""\
送料は{POSTAGE_FEE}円です
送料込税込価格は{total_price}円です""")

```
