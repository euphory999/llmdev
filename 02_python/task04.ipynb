```python
try:
    result = int(input("整数 a を入力してください: ")) / int(input("整数 b を入力してください: "))

# 0で割る場合の例外の捕捉
except ZeroDivisionError:
    print("エラー: 0で割ることはできません")

# 不正値の例外の捕捉
except ValueError:
    print("エラー: 整数を入力してください")

# 未知の例外の捕捉
except Exception:
    print("エラー: なんらかの例外が発生しました")

# 上で捕捉している例外が起きなかったら通る
else:
    print(f"計算結果: {result}")

# 例外が起きても起きなくても必ず通る
finally:
    print("プログラム終了")
```
