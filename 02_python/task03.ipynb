````python
import re

# 1. 元金 (base) の入力
is_validation_ng = True
while is_validation_ng:
    raw_base = input("元金を入力してください: ")
    is_validation_ng = re.match(r"^[0-9]+$", raw_base) is None
    if is_validation_ng:
        print("元金は全て半角数字で入力してください!")
base = int(raw_base)

# 2. 年利 (cmpl_rate) の入力
while True:
    raw_cmpl = input("年利を入力してください: ")
    try:
        cmpl_rate = float(raw_cmpl)
        break
    except:
        print("元金は小数に変換できる値を入力してください! (例:5%と入力したいならば5.00と入力してください)")

# 3. 10年の複利計算 (今回は回数が分かっているので for を使う)
# Note. 金額の精度や端数については今回考慮していません
result = base
for _ in range(10):
    result *= (100.0 + cmpl_rate)/100.0

# 4. 計算結果の表示
print(f"10年後の金額: {result:.2f}")
````

もしも単純に $ x^y $ を求めたい場合は `**` 演算子も使える

```python
rate = 1.05 ** 10

# -> 1628.894626777442
print(1000 * rate)
```

ただし演算順序には気をつけたい
指数演算の右項優先は数学上の定義でもある

```python
# -> 2 (2^1 を先に計算して残った 2^4ではない点に注意したい)
2 ** 1 ** 4
```
