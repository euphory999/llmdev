# VS Code で Jupyter を利用する

## 1. venv で仮想環境を構築して有効化する

```
$ python -m venv venv
$ ./venv/bin/activate
```

## 2. 仮想環境に jupyter をインストールする

※ もしかしたら `ipykernel` だけでもよいかもしれません

```
$ pip install jupyter==1.1.1
```

## 3. VSCode の Kernel として venv のパスを選択する

右上にある Kernel 選択画面で venv のパスを選択することで jyputer を起動することができる
