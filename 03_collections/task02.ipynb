```python
import re

# 1. 英単語の入力フェーズ
words = []
chars = {}

while True:
    word = input("英単語を入力してください：")
    if (m := re.match(r"^[a-zA-Z]*$", word)) is not None:
        # 正常系
        if word == "":
            # 空文字なら終了
            break
        else:
            # 何らかの単語ならば追加
            words.append(word)

            # 単語の文字を辞書に追加していく
            for char in word:
                # 'p' という文字を追加する場合 chars['p'] が存在するかどうかをチェックします
                # - もし存在しない場合は chars['p'] = 1 として新しいキーとともに初期値を定義します
                # - 存在する場合は chars['p'] += 1 します
                # これにより文字数をカウントします
                chars[char] = chars[char] + 1 if char in chars else 1

    else:
        # 異常系のハンドル
        print("a-z, A-Z のみで構成される英単語を入力してください (空文字で入力を終了します)")

# 2. 英単語のリストをABC順に並べ替えて一覧表示
print(f"入力した英単語:{sorted(words)}")

# 3. 文字のカウント数を順に表示していく
for char in sorted(chars.keys()):
    print(f"{char}が{chars[char]}個ありました")
```
